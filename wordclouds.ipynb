Importing the data and first look.

```python
df = pd.read_csv('sentiment_results1.csv')
```

Splitting the data into positive and negative results

```python
# subset result = Positive
df_pos = df[df['result'] == 'Positive']
```

```python
# subset result = Negative
df_neg = df[df['result'] == 'Negative']
```

Ploting the wordclouds
With the folowing instructions:
This code snippet is a Python script for generating and displaying a word cloud from a set of texts. Here's a breakdown of its functionality:

1. **Text Aggregation**:
   - `text = ' '.join(df_pos['text'].astype(str))`: 
     - This line takes a DataFrame `df_pos` and selects the column named `'text'`.
     - Each element in this column is converted to a string (if not already a string) using `astype(str)`.
     - All these strings are then joined together into a single long string, separated by spaces. This is done by `' '.join(...)`.
     - The resulting string, `text`, contains all the text data from the `'text'` column of `df_pos`.

2. **Word Cloud Creation**:
   - `wordcloud = WordCloud(width = 1000, height = 500).generate(text)`:
     - A `WordCloud` object is created with a specified width of 1000 pixels and height of 500 pixels.
     - The `generate(text)` method is called on this object, using the aggregated text. This method creates a word cloud from the provided text, where the size of each word in the cloud is proportional to its frequency in the text.

3. **Displaying the Word Cloud**:
   - `plt.figure(figsize=(15,8))`:
     - Sets up a matplotlib figure with a specified size (15x8 inches) to display the word cloud.
   - `plt.imshow(wordcloud, interpolation='bilinear')`:
     - Displays the generated word cloud image. The `interpolation='bilinear'` argument smooths the displayed image.
   - `plt.title('Word Cloud for offensive language', fontsize=20)`:
     - Adds a title to the figure with a font size of 20.
   - `plt.axis("off")`:
     - Turns off the axis lines and labels.
   - `plt.show()`:
     - Renders the figure and displays it to the user.

The overall purpose of this script is to visualize the frequency of words in the text data from `df_pos`, specifically targeting texts that may contain offensive language (as suggested by the title in the plot). The word cloud visually emphasizes more frequent words by making them larger.
- 

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all text into a single string
# Combine all text into a single string
text = ' '.join(df_pos['text'].astype(str))

# Create a WordCloud object
wordcloud = WordCloud(width = 1000, height = 500).generate(text)

# Display the generated image
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud for  ofensive language', fontsize=20)
plt.axis("off")
plt.show()

wordcloud.to_file("wordcloud_pos.png")
```

```python
# Combine all text into a single string
text_neg = ' '.join(df_neg['text'].astype(str))

# Create a WordCloud object
wordcloud_neg = WordCloud(width = 1000, height = 500).generate(text_neg)

# Display the generated image
plt.figure(figsize=(15,8))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.title('Word Cloud for non   ofensive language', fontsize=20)
plt.axis("off")
plt.show()

# save image
wordcloud_neg.to_file("wordcloud_neg.png")
```
